```python
def convert(impath_train_label,impath_train_img):
    
    df=pd.DataFrame([],columns=['filename','width','height','class','xmin','ymin','xmax','ymax'])
    
    for file in sorted(os.listdir(impath_train_label)):
        #print(file)
        try:
            content=pd.read_csv(os.path.join(impath_train_label,file),header=None)

            img=cv2.imread(os.path.join(impath_train_img,file.replace('txt','jpeg')))
            #print(img)
            for i in range(1,pd.to_numeric(content.iloc[0][0])+1):
                
                
    
                temp_df=pd.DataFrame({'filename': [file.replace('txt','jpeg')],'width':[img.shape[1]],'height':[img.shape[0]],
                                      'class':['gun'],'xmin':[pd.to_numeric(str.split(content.iloc[i][0]))[0]],
                                      'ymin':[pd.to_numeric(str.split(content.iloc[i][0]))[1]],
                                      'xmax':[pd.to_numeric(str.split(content.iloc[i][0]))[2]],
                                      'ymax':[pd.to_numeric(str.split(content.iloc[i][0]))[3]]})
                
                df=df.append(temp_df)
                print(temp_df)
                #print(pd.to_numeric(content.iloc[0][0]))
        except:
            print(file)
    return df.reset_index(drop=True)
```

```python
ff=convert(impath_train_label,impath_train_img)
ff.to_csv('/home/lenovo/anaconda3/lib/python3.7/site-packages/tensorflow/workspace/training_demo/annotations/train_labels.csv',index=False)
```

```python
gg=convert(impath_test_label,impath_test_img)
gg.to_csv('/home/lenovo/anaconda3/lib/python3.7/site-packages/tensorflow/workspace/training_demo/annotations/test_labels.csv',index=False)
```
